## ***11. Drop The Columns education-num, capital-gain and capital-loss***

## ***12. What Is The Distribution of Age Column Show In Graph?***


```python
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## ***13. Find Total Number of Persons Having Age Between 17 To 48 (Inclusive) Using Between Method And Also Without Using Between Method***

```python
age_between = df_cleaned[df_cleaned['age'].between(17, 48)].shape[0]
age_between
```

```python
age_without_between = df_cleaned[(df_cleaned['age'] >= 17) & (df_cleaned['age'] <= 48)].shape[0]
age_without_between
```
